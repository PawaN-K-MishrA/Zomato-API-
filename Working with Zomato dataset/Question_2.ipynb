```python
#importing the requred libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Solution to Question 2 part 1(a)

```python
#reading the dataset from the file and creating the datafrmae.
data=pd.read_csv('zomato.csv',encoding='latin-1')

#copying the data to avoid changes in orignal datafrmae.
df=data.copy()

#selecting the restaurants of country==1 (i.e, INDIA).
df1=df[df['Country Code']==1]

#creating the rating array using np.array()
rating=np.array(df1['Aggregate rating'],dtype=float)

#creating the votes array using np.array()
votes=np.array(df1['Votes'])

#plotting the graph.
plt.figure(figsize=(15,7))
plt.scatter(rating,votes,marker='^',color='blue',alpha=0.3)
plt.xticks(fontsize=12,fontweight='bold')
plt.yticks(fontsize=12,fontweight='bold')
plt.ylabel('Number of votes------->',fontsize=15,fontweight='bold')
plt.xlabel('Aggregate rating------>',fontsize=15,fontweight='bold')
plt.title('Number of Votes with Aggregate rating',fontsize=20,fontweight='bold')
plt.grid()
plt.show()
```

# Analysis:
Clearly as the number of votes increases the chances of getting higher ratings also increases. It is clear from the graph that any restaurants with more than 5000 votes get ratings between 4 and 5. when the voes are less than 2000, there is a diversity of ratings, which varies from 2 to 5. so for a restaurant to have a probability of getting higher aggregate rating, it should have more than 5000 votes. Also there are many restaurants having votes between between 1000 and 2000 with very good ratings, that is between 4 and 5.

# Solution to Question 2 part 1(b)

```python
#reading the dataset from the file and creating the datafrmae.
data=pd.read_csv('zomato.csv',encoding='latin-1')

#copying the data to avoid changes in orignal datafrmae.
df=data.copy()

#selecting the restaurants of country==1 (i.e, INDIA).
df=df[df['Country Code']==1]

#creating the list to determine the number of cuisnes served by restaurants.
cuisine_offered=[]

#itearting over cuisines values.
for i in df['Cuisines'].values:
    
    #getting the count of cuisines offered and appending them in cuisine_offered.
    cuisine_offered.append(len(i.split(',')))
    
#creating the new column for number of cuisines offered by each restaurant.
df['cuisine_count']=cuisine_offered

#creating the array of number of cuisines offered.
count_arr=np.array(df['cuisine_count'])

#creating the dictionary to store the number of restaurants offering by number of cusines as key value pair. 
count_dict={}

#iterating over count_arr.
for i in count_arr:
    
    #increasing the resaturant count of number of cuisnes.
    count_dict[i]=count_dict.get(i,0)+1

#creating the rating array.
rating_arr=np.array(df['Aggregate rating'],dtype=float)

#getting the values of number of restaurants number of cuisines.
count=list(count_dict.values())

#printing the table.
print('Number of restaurants serving the provided number of cuisines.:')
print('+-----------------------------------------------------------------------------------------+')
print('|'+' '*3 +'Number of Cuisines served'+' '*7+'|'+' '+'No. of restaurants serving given number of cuisines'+' '+'|')
print('+-----------------------------------------------------------------------------------------+')
for i in count_dict.keys():
    print('|'+' '*2+str(i)+' '*(32-len(str(i))),'|'+' '*5+str(count_dict[i])+' '*(47-len(str(count_dict[i])))+' '+'|')
print('+-----------------------------------------------------------------------------------------+')

#plotting the graph.
plt.figure(figsize=(15,5))
plt.scatter(rating_arr,count_arr,color='orange',alpha=0.5,marker='s')
plt.xticks(fontsize=12,fontweight='bold')
plt.yticks(fontsize=12,fontweight='bold')
plt.ylabel('Number of Cuisines------->',fontsize=15,fontweight='bold')
plt.xlabel('Aggregate rating------>',fontsize=15,fontweight='bold')
plt.title('Number of Cuisines with Aggregate rating.',fontsize=20,fontweight='bold')
plt.grid()
plt.show()
```

# Analysis:
From the graph itself, it is clear that when the number of cuisines provided increases from 3 to 8, generally the rating seems to converge between 3 and 4. restaurants providing more number of cuisines are not much likely to get higher ratings, specially when the number of cuisines provided exceeds 6. while if the restaurants provide less number of cuisines, for example 1 or 2 cuisines, they are more likely to get higher ratings. Infact many restaurants in india are providing 2 cuisines. there is a very less number of restaurants providing 8 cuisines. it seems like when a restaurant provides too many cuisines, its focus on the quality of food offered diverges. while restaurants providing less cuisines focus on the quality of food to get good aggregate ratings.

# Solution to Question 2 part 1(d)

```python
#reading the dataset from the file and creating the datafrmae.
data=pd.read_csv('zomato.csv',encoding='latin-1')

#copying the data to avoid changes in orignal datafrmae.
df=data.copy()

#selecting the restaurants of country==1 (i.e, INDIA).
df=df[df['Country Code']==1]

#creating the rating array.
rating=np.array(df['Aggregate rating'],dtype=float)

#creating the cuisines array.
cuisines=np.array(df['Cuisines'])

#creating the cuisine_count dictionary.
cuisine_count={}

#itearting over cuisines.
for i in range(len(cuisines)):
    
    #multiple cuisines are served by restaurants splitting them using delim ','.
    for cus in cuisines[i].split(','):
        
        #storing the count for each cuisine by using .strip() method to avoid spaces.
        cuisine_count[cus.strip()]=cuisine_count.get(cus.strip(),0)+1
        
#sorting the cuisine_sorted dictionary in decreasing order based upon values to get top 15.
cuisine_sorted={k:v for k,v in sorted(cuisine_count.items(),key=lambda item:item[1],reverse=True)}

#getting the name of top 15 cuisines.
top_15_cuisines=list(cuisine_sorted.keys())[:15]

#creating the avg_rating array to get rating for each cuisine.
avg_rating=[]

#itearting over the top_15_cuisines.
for i in top_15_cuisines:
    sum1,count=0,0
    
    #itearating over rating column to get avg for each cuisines.
    for j in range(len(rating)):
        if i in cuisines[j]:
            sum1+=rating[j]
            count+=1
            
    #appending the avearge of each cusine in avg_rating.
    avg_rating.append((sum1/count))
    
#creating the colour array.
color_arr=['#16A085','#FBEEE6','#ABB2B9','#E6B0AA','#D2B4DE','#AED6F1','#F9E79F','#FAD7A0','#EDBB99','#E5E7E9',
          '#C39BD3','#7FB3D5','#85929E','#2C3E50','#C0392B']

#plotting the graph.
plt.figure(figsize=(18,5))
plt.bar(top_15_cuisines,avg_rating,color=color_arr,width=0.5)
plt.xticks(fontsize=10,fontweight='bold')
plt.yticks(fontsize=14,fontweight='bold')
plt.ylabel('Aggregate rating------->',fontsize=14,fontweight='bold')
plt.xlabel('Cuisines------>',fontsize=14,fontweight='bold')
plt.title('Aggregate rating of Cuisines',fontsize=14,fontweight='bold')
plt.grid()
plt.show()

    
```

# Analysis:
These are top 15 cuisines that served maximum number of time all the restaurant. Calculated average rating for top cuisines and most of the cuisines having good average rating and italian cuisines is at top. From above graph we can see restaturant serving Cuisine 'Italian','Continental','American' have more average rating compared to other cuisine

# Solution to Question 2 part 2

```python
#reading the dataset from the file and creating the datafrmae.
da=pd.read_csv('zomato.csv',encoding='latin-1')

#copying the data to avoid changes in orignal datafrmae.
df=da.copy()

#selecting the restaurants of country==1 (i.e, INDIA).
df=df[df['Country Code']==1]

#converting the list of tuples where each tuple contain the cuisines,votes and rating.
data=list(zip(df['Locality'],df['Votes'],df['Aggregate rating']))

#creating the locality array.
locality=df['Locality'].value_counts().values

#itearting over locality and getting the index of top 50 elements.
for i in range(len(locality)):
    if locality[i]<50:
        index=i
        break

#getting the top 50 localities.
locality=df['Locality'].value_counts().index[0:index+1]

#creating the weight_list to append the weighted restaurant rating.
weight_list=[]
for i in locality[:10]:
    N_votes=0
    rat=0
    
    #itearating over data list.
    for j in range(len(data)):
        if i in data[j][0]:
            rat=rat+(data[j][1]*data[j][2])
            N_votes=N_votes+data[j][1]
    weight_list.append(rat/N_votes)

#printing the table.
print('+------------------------+')
print('|'+' '*(5)+'Locations'+' '*(10)+'|')
print('+------------------------+')

#itearting over the returned set.
for i in range(10):
    
    #printing the cuisines.
    print('|'+' '*2+str(locality[i])+' '*(22-(len(str(locality[i]))))+'|')
print('+------------------------+')
    
#declaring the color_arr.
color_arr=['#BFFC01','#700E61','#FA033B','#A5475C','#352B2D','#68718B','#2CAFAB','#FEFA00','#4C5A91','#E700FE']

#plotting the graph.
plt.figure(figsize=(20,7))
plt.bar(locality[:10],weight_list,color=color_arr[::-1],width=0.4)
plt.xticks(fontsize=12,fontweight='bold')
plt.yticks(fontsize=12,fontweight='bold')
plt.ylabel('Weighted Restaurant rating------->',fontsize=15,fontweight='bold')
plt.xlabel('Localities------>',fontsize=15,fontweight='bold')
plt.title('Top 10 localities with Weighted Restaurant rating ',fontsize=20,fontweight='bold')
plt.grid()
plt.show()
            
```
