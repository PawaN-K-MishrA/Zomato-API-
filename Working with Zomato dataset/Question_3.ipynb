# Solution to Question 3 part(1)

```python
#reading the dataset from the file and creating the datafrmae.
da=pd.read_csv('zomato.csv',encoding='latin-1')

#copying the data to avoid changes in orignal datafrmae.
df=da.copy()

#selecting the restaurants of country==1 (i.e, INDIA).
df=df[df['Country Code']==1]

#creating the restaurant array.
restaurants=np.array(df['Restaurant Name'])

#creating the dictionary to store number of outlets as key value pair.
restaurant_outlets={}

#itearating over restaurant list.
for i in range(len(restaurants)):
    
    #increamenting the value of count of outlet for restaurants.
    restaurant_outlets[restaurants[i]]=restaurant_outlets.get(restaurants[i],0)+1
    
#sorting the restaurant_outlets dictionary in descrasing order.
restaurant_outlets_sorted={k:v for k,v in sorted(restaurant_outlets.items(),key=lambda item:item[1],reverse=True)}

#getting the list of top 15 restaurants.
restaurants_15=list(restaurant_outlets_sorted.keys())[:15]

#getting the number of oulets for top 15 restaurants.
counts=list(restaurant_outlets_sorted.values())[:15]

#plotting the graphs.
plt.figure(figsize=(12,8))
plt.barh(restaurants_15[::-1],counts[::-1],color=color_arr[:15])
plt.xticks(np.arange(0,90,5),fontsize=12,fontweight='bold')
plt.yticks(fontsize=12,fontweight='bold')
plt.xlabel('Count of Outlets------->',fontsize=15,fontweight='bold')
plt.ylabel('Restaurants in increasing order of outlets------->',fontsize=15,fontweight='bold')
plt.title('Top 15 Restaurants with maximum outlets.',fontsize=20,fontweight='bold')
plt.grid()
plt.show()
```

# Solution to Question 3 part(3)

```python
#reading the dataset from the file and creating the datafrmae.
da=pd.read_csv('zomato.csv',encoding='latin-1')

#copying the data to avoid changes in orignal datafrmae.
df=da.copy()

#selecting the restaurants of country==1 (i.e, INDIA).
df=df[df['Country Code']==1]

#sorting the dataframe based upon votes column in descending order
df=df.sort_values(by=['Votes'],ascending=False)

#declaring lists and votes list.
restaurant,votes=[],[]

#itearating over the dataframe values list using .values.
for i in df.values[:11]:
    
    #appending the name of restaurants which is present at index 1 of each row.
    restaurant.append(i[1])
    
    #appending the count of votes which is present at index 20 of each row.
    votes.append(i[20])

#plotting the horizontal bar graph.
plt.figure(figsize=(12,5))
plt.barh(restaurant[::-1],votes[::-1],color=color_arr[::-1][:11])
plt.xticks(fontsize=12,fontweight='bold')
plt.yticks(fontsize=12,fontweight='bold')
plt.xlabel('Votes---------->',fontsize=15,fontweight='bold')
plt.ylabel('Restaurant Name------->',fontsize=15,fontweight='bold')
plt.title('Restaurants with Votes',fontsize=20,fontweight='bold')
plt.grid()
plt.show()
```

# Solution to Question 3 part(4)

```python
#reading the dataset from the file and creating the datafrmae.
da=pd.read_csv('zomato.csv',encoding='latin-1')

#copying the data to avoid changes in orignal datafrmae.
df=da.copy()

#selecting the restaurants of country==216 (i.e, USA).
df=df[df['Country Code']==216]

#dropping the null values of cuisines column of country USA.
df['Cuisines'].dropna(inplace=True)

#declaring dictionary to store the count of each cuisines as key value pair.
cuisine_dict={}

#iteartin gover the cuisines column values using .value .
for i in df['Cuisines'].values:
    
    #since there are multiple cuisines served so splitting them with delim as ','.
    for j in i.split(','):
        
        #avoiding extra space using .strip() and adding the count of each cuisines.
        cuisine_dict[j.strip()]=cuisine_dict.get(j.strip(),0)+1
        
#sorting the dictionary based upon value in descreasing order.
cuisine_dict_soretd={k:v for k,v in sorted(cuisine_dict.items(),key=lambda item:item[1],reverse=True)}

#Getting the name of top 10 cuisines.
cusine_name=list(cuisine_dict_soretd.keys())[:10]

#getting the count of top 10 cuisines.
count=list(cuisine_dict_soretd.values())[:10]

#plotting the pie chart.
legend=[]
for i in cusine_name:
    legend.append(i)
plt.pie(count,autopct='%.2f%%',radius=3.5,colors=color_arr[10:21])
plt.title('Pie Chart for Top 10 Cuisines of USA',fontdict={'fontsize':20})
plt.legend(legend,bbox_to_anchor=(0,1))
plt.axis('equal')
plt.show()
```

# Solution to Question 3 part(5)

```python
#reading the dataset from the file and creating the datafrmae.
da=pd.read_csv('zomato.csv',encoding='latin-1')

#copying the data to avoid changes in orignal datafrmae.
df=da.copy()

#selecting the restaurants of country==1 (i.e, INDIA).
df=df[df['Country Code']==1]

#getting the count of restaurants in cities of india using .value_counts().
nf=df['City'].value_counts()

#creating the array to store the names of cities.
city=np.array(nf.index)

#decalring the array to store the aggregate weight.
aggregate_=[]

#creating the array to store the counts of restaurants in cities of india.
count=np.array(nf.values)

#iterating to the length of city columns.
for i in range(len(city)):
    
    #creating the datafrmae for the selected city to get details of restaurants in that particular city.
    new_df=df[df['City']==city[i]].copy()
    
    #Adding the column rate in above created dataframe.
    new_df['rate']=new_df['Aggregate rating']*new_df['Votes']
    
    #getting the aggregate weight.
    total=new_df['rate'].sum()
    count_v=new_df['Votes'].sum()
    aggregate_.append((total/count_v))
    
#Rounding the values of aagregate weight upto 1 decimal place.
aggregate_=[round(i,1) for i in aggregate_]

#plotting the graph.
plt.subplots(figsize=(25,10))
plt.scatter(city,count,s=350,edgecolors='black',alpha=1,color='orange')
plt.xticks(rotation=90,fontsize=20,fontweight='bold')
plt.yticks(fontsize=20,fontweight='bold')
plt.xlabel('Cities---------->',fontsize=25,fontweight='bold')
plt.ylabel('count of restaurants----->',fontsize=25,fontweight='bold')
plt.title('Restaurants in city with aggregate rating.',fontsize=25,fontweight='bold')
for i in range(len(city)):
    plt.text(city[i],count[i],aggregate_[i],fontweight='bold')
plt.grid()
plt.show()

    
        
        
  
   
    

    
```
